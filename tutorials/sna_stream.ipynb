# Social Network Analysis of Live Streaming Data

A simple way of working with hot topics is to plot them in real time. You can find trends on <https://twitter.com/>, or research words you're interested in. But since Twitter's public API streams only 5% or so tweets meeting the criteria, the stream for not-so-hot trends gets new tweets at the rate below 10 tweet per second. That's kinda slow for real time research. I recommend making `json` dumps in that case.

But new tweets arrive fast for hot trends. Here's one example of working with them in real time:

```python
%pylab inline

import twitternodes as tn
TWORD = "#MTVEMA"
```

Variable `TWORD` contains the words we're looking for in the Twitter stream. [This page](https://dev.twitter.com/streaming/overview/request-parameters#track) explains how keywords work. Actually, you can use all the parameters allowed in [Streaming API request parameters](https://dev.twitter.com/streaming/overview/request-parameters), but that's for another tutorial.

Let's plot the recent 100 plots related to the keyword:

```python
tn.plot_live(n=100, track=TWORD)
```

This is a graph for tags. The graph for user mentions associated with the tag looks like this:

```python
tn.plot_live(type="user_mentions", n=100, track=TWORD)
```

The time span of real-time data is small, so the sample is not very representative. But you can notice cliques around certain users.

The data is also available in a `pandas` dataframe:

```python
db = tn.DB("hashtags")
db.populate(tn.Stream().fetch_live(n=100, track=TWORD))
nodes = db.group_nodes()
edges = db.weigh_edges()
```

The four lines create a new object, populate it with live data, and generate two summary dataframes for nodes and edges.

The weight for edges equals the number of mentions of two tags:

```python
edges.sort('weight', ascending=False)[:10]
```

And further you can work with the data as in [sna_for_journalists_1.ipynb](http://nbviewer.ipython.org/github/antontarasenko/twitter-nodes/blob/master/tutorials/sna_for_journalists_1.ipynb). To convert the data into a graph:

```python
G = tn.graph(nodes, edges)
```

Which is a `networkx` object, well documented [here](http://networkx.github.io/documentation/latest/overview.html).

# Social Network Analysis for Journalists: Introduction

# Introduction

Researching a topic is difficult. Social media help, but on the way they create much noise, so you need to filter it somehow. Here's a few tricks that can make research in social media easier (or more difficult, depending on how long you practice these tricks).

The approach relies on graph theory. It's from math, but the concepts is so intuitive that you won't recall this word again.

Examples are based on Python package called [Twitter Nodes](https://github.com/antontarasenko/twitter-nodes), which I wrote for handling tweets. You can use any other tools as far as they can do what you'll see below.

# Getting Started

## Bird-eye View on the Topic

Let's start with drawing a _concept map_ for the recent [Ferguson events](http://en.wikipedia.org/wiki/2014_Ferguson_unrest). What happened in Ferguson is a tragedy. Discrimination in general is a sensitive topic, so it's important to have objective information.

The concept map would include all topics mentioned on Twitter together with Ferguson. Technically, I extract all hashtags and user mentions that appear in tweets containing word "ferguson". I obtained 1,500 of such tweets. They're in the file called __ferguson-1500.json__, which I load immediately:

```python
import twitternodes as tn
tags = tn.DB("hashtags")
nodes, edges = tags.load_json("data/ferguson-1500.json")
```

Now I draw what I have in this file:

```python
G = tn.graph(nodes, edges)
tn.plot(G)
```

What a mess! This is how Twitter looks like as is.

Now we'll make it clearer. First, we have two tables called `nodes` and `edges`. Nodes are tags and edges are connections between tags. Each connection is two tags mentioned together in a single tweet. As the graph above says, __ferguson-1500.json__ contains 386 tags and 881 connections between them. This is too much, and I'll filter them by the number of times tags have been mentioned in the sample:

```python
important_nodes = nodes[nodes['mentions'] > 5]
```

That's standard [pandas'](http://en.wikipedia.org/wiki/Pandas_%28software%29) way to select the data. Let's plot the graph now:

```python
iG = tn.graph(important_nodes, edges)
tn.plot(iG, figsize=(15, 10))
```

It's easier to understand the plot now. A line's thickness shows the number of times two tags were mentioned together. The radius of the circle reflects the number of tag mentions.

Have a look at the same results in a table:

```python
edges.sort('weight', ascending=False)[:10]
```

This table shows most mentioned pairs of tags. Tag "tcot" stands for "Top Conservatives on Twitter." Twitter users connect it with racism in America. They also demand to arrest Darren Wilson, who shot Mike Brown. For some users, it would mean justice for Mike Brown, as you see from the ninth row.

Now, most mentioned tags:

```python
important_nodes.sort('mentions', ascending=False)[:10]
```

Conservatives again get mentioned more frequently than other topics. Ferguson itself is mentioned so frequently because it was the keyword for filtering the Twitter stream. So, _tag_ "ferguson" also appears in most tweets.

Facebook is a popular tag because people organize in that social network too, so cross-posting affects the results.

The same in a graph:

```python
sorted_nodes = nodes.sort('mentions')[-20:-1]
sorted_nodes.index = sorted_nodes.name
sorted_nodes['mentions'].plot(title="Tag mentions", kind='barh')
```

# Network Analysis

Twitter Nodes supports all operations implemented in [networkx](https://github.com/networkx/networkx) package. To illustrate some applications, I focus on the big sample created above.

## Cliques

```python
import networkx as nx
for i in nx.find_cliques(G):
    if len(i) > 7: print(i)
```

This command selects several large [_cliques_](http://en.wikipedia.org/wiki/Clique_%28graph_theory%29). A clique here is a collection of tags that have been mentioned with each other at least once. It's better to think about cliques as topics.

The topics give hints for further research. For example, "cyberattacks" appear because the FBI investigates the case of cyberattacks on Ferguson law enforcement. "Civil disobedience" emerges alongside with "ferguson shooting." Topics change over time, and a newer sample of tweets would give an updated list of cliques.

## Centrality

Since it's all about connections, some tags may be important because they happen to be in the center of these connections:

```python
tn.plot(iG, figsize=(15,10))
```

We can calculate [centrality](http://en.wikipedia.org/wiki/Centrality) like this:

```python
import pandas as pd
centrality = pd.DataFrame(list(nx.degree_centrality(G).items()), columns=['name', 'centrality'])
centrality.set_index('name').sort('centrality')['centrality'][-20:-1].plot(kind='barh')
```

[Degree centrality](http://en.wikipedia.org/wiki/Degree_%28graph_theory%29) above is based the number of connections each tag has with other tags.

It contributes a little to what we already know from previous tables, and I'll use it for another case.

# Researching Users

The previous sections dealt with hashtags. You can use the same code for learning about user mentions. Just replace `tn.DB("hashtags")` with `tn.DB("user_mentions")` in the very first input line and re-run the code.

Otherwise, let's try something different.

```python
users = tn.DB("user_mentions")
U = tn.graph(*users.load_json("data/ferguson-1500.json"))
```

## Importance

When two users are mentioned together in the same tweet, users with many connections end up being more important than others. That's degree centrality we measured for tags. You can try that.

But here's another measure: [betweenness centrality](http://en.wikipedia.org/wiki/Betweenness_centrality), which shows the number of shortest paths coming through a given node. Here's a good comparison of different centrality measures from Wikipedia:

![Centrality](http://upload.wikimedia.org/wikipedia/commons/thumb/1/14/Centrality.svg/300px-Centrality.svg.png)

A) Degree centrality, B) Closeness centrality, C) Betweenness centrality, D) Eigenvector centrality, E) Katz centrality and F) Alpha centrality of the same graph.

Calculating betweenness:

```python
user_betweenness = pd.DataFrame(list(nx.betweenness_centrality(U).items()), columns=['name', 'betweenness'])
user_betweenness.set_index('name').sort('betweenness')['betweenness'][-20:-1].plot(kind='barh')
```

A puzzling leader is [@tip](http://twitter.com/tip). That's because [T.I.P.](http://en.wikipedia.org/wiki/T.I.) released a single in support for Ferguson.

User [@clairecmc](https://twitter.com/clairecmc) is a Missouri senator involved in mitigating the consequences of the events.

Mass media have relatively high betweenness, so they're the nodes that connect other mentioned users. But personal accounts, like @blogdiva and @talibkweli, are in the top as well. So, they play the role of the media as well.

## Neighbors

You can learn a lot about public figures by looking at who they're mentioned with. With graphs, it's simple:

```python
U.neighbors('clairecmc')
```

The senator is mentioned together with the media, mostly. That implies she isn't oft-mentioned with ordinary users. Otherwise, we would expect tweets like ".@clairecmc visits @citizen", and then @citizen would enter the list above.

For comparison, the similar list of "neighbors" for @blogdiva:

```python
U.neighbors('blogdiva')
```

That's quite a different result. You can explore these connections as well.
